## **Intregração Python com o SQL**

O primeiro passo é realizar a importação da biblioteca **pyodbc**

Após a implementação dessa biblioteca, devemos seguir três caminhos:

- Criar dados de conexão (data_conection)

        Driver: Qual SGBD estamos utilizando.
        Servidor: Local ou da empresa (para descobrirmos o servidor local, basta digitar no prompt: hostname).
        DataBase: Nome do banco de dados.

- Criar a conexão

        pyodbc.connect(data_conection)
        
- Importar o banco de dados através do pandas

```python
import pyodbc
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

```python
data_conection = ("Driver={SQL Server};"
                 "Server=DESKTOP-0EKP870;"
                 "Database=ContosoRetailDW;")
```

----------------
### **Desafio**

Descobrir o lucro diário de uma empresa

        Lucro = Vendas - Custo - Descontos 

**Observação:** Em um dia, podem existir mais de uma transação. Então é necessário agrupar od dados registrados no mesmo dia. 

--------------
**Conexão + leitura**

Necessário escrever o código em SQL e depois fazer a importação de quais tabelas desejamos importar e manipular 

```python
conection = pyodbc.connect(data_conection)
```

```python
sql_code = """

SELECT 
    SalesKey as Id,
    DateKey as Datas,
    DiscountAmount as Descontos,
    TotalCost as 'Total de custos',
    SalesAmount as 'Quantidade vendas'
FROM
    ContosoRetailDW.dbo.FactSales
"""
sell_of_company = pd.read_sql(sql_code, conection)
```

```python
display(sell_of_company)
```

------------
**Tratamento de dados**

```python
sell_of_company.info()
```

----------------
**Criando uma coluna de vendas**

```python
sell_of_company['Lucros'] = (sell_of_company['Quantidade vendas'] - 
                             sell_of_company['Descontos'] - sell_of_company['Total de custos'])
display(sell_of_company)
```

------------
**Agrupamento dos dados**

nome_da_tabela.groupby(['nome_coluna_referência']).operação() 

```python
day_profit = sell_of_company.groupby('Datas').sum()
display(day_profit)
```

----------------
**Visualização dos dados**

```python
graph = day_profit['Lucros'].plot(figsize=(15,5), color='blue')
graph.yaxis.set_major_formatter(matplotlib.ticker.StrMethodFormatter('${x:,.0f}'))
plt.grid(linestyle='--')
plt.show()
```

----------
