```python
import pandas as pd
import random 
import numpy as np
import string
from random import shuffle
```

```python
lower_alphabet = string.ascii_lowercase
```

```python
all_prices = []
for i in range(len(lower_alphabet)):
    prices = [random.uniform(1,100) for j in range(365)]
    all_prices.append(prices)
```

```python
data_frame = pd.DataFrame(np.array(all_prices).T, columns=list(lower_alphabet))
```
